```python
import random


data_list = labels_

train_ratio = 0.8
dev_ratio = 0.1
test_ratio = 0.1

total_elements = len(data_list)
train_count = int(total_elements * train_ratio)
dev_count = int(total_elements * dev_ratio)
test_count = int(total_elements * test_ratio)


if total_elements != train_count + dev_count + test_count:
    extra_for_train = train_count - (total_elements - dev_count - test_count)
    train_count += extra_for_train

random.shuffle(data_list)


def write_to_file(filename, elements, count):
    with open(filename, 'w', encoding='utf-8') as file:
        for element in elements[:count]:
            file.write(element + '\n')

# ÂàÜÂà´ÂÜôÂÖ•ËÆ≠ÁªÉ„ÄÅÂºÄÂèëÂíåÊµãËØïÊñá‰ª∂
write_to_file('train.label', data_list[:train_count], train_count)
write_to_file('dev.label', data_list[train_count:train_count + dev_count], dev_count)
write_to_file('test.label', data_list[train_count + dev_count:], test_count)
```

```python
import random

data_list = texts

# Á°ÆÂÆöÊØè‰∏™Êñá‰ª∂Â∫îÂåÖÂê´ÁöÑÂÖÉÁ¥†Êï∞Èáè
train_ratio = 0.8
dev_ratio = 0.1
test_ratio = 0.1

# ËÆ°ÁÆóÊØè‰∏™Êñá‰ª∂Â∫îÂåÖÂê´ÁöÑÂÖÉÁ¥†Êï∞Èáè
total_elements = len(data_list)
train_count = int(total_elements * train_ratio)
dev_count = int(total_elements * dev_ratio)
test_count = int(total_elements * test_ratio)

# Â¶ÇÊûúÂÖÉÁ¥†ÊÄªÊï∞‰∏çËÉΩË¢´ÊØî‰æãÊï¥Èô§ÔºåÊàë‰ª¨ÈúÄË¶Å‰ªéËÆ≠ÁªÉÈõÜ‰∏≠ÊäΩÂèñÈ¢ùÂ§ñÁöÑÂÖÉÁ¥†‰ª•Á°Æ‰øùÊØî‰æãÊ≠£Á°Æ
if total_elements != train_count + dev_count + test_count:
    extra_for_train = train_count - (total_elements - dev_count - test_count)
    train_count += extra_for_train

# Êâì‰π±ÂàóË°®ÁöÑÈ°∫Â∫è
random.shuffle(data_list)

# ÂÜôÂÖ•Êñá‰ª∂
def write_to_file(filename, elements, count):
    with open(filename, 'w', encoding='utf-8') as file:
        for element in elements[:count]:
            # ÁßªÈô§ÂÖÉÁ¥†‰∏≠Èó¥ÁöÑÊç¢Ë°åÁ¨¶
            cleaned_element = ''.join(element.split('\n'))
            file.write(cleaned_element+"\n")

# ÂàÜÂà´ÂÜôÂÖ•ËÆ≠ÁªÉ„ÄÅÂºÄÂèëÂíåÊµãËØïÊñá‰ª∂
write_to_file('train.txt', data_list, train_count)
write_to_file('dev.txt', data_list[train_count:train_count + dev_count], dev_count)
write_to_file('test.txt', data_list[train_count + dev_count:], test_count)
```
